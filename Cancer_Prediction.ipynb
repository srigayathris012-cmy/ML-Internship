```python
ind = dia[['Age', 'Gender', 'BMI', 'Smoking','GeneticRisk','PhysicalActivity','AlcoholIntake','CancerHistory']]   # independent variables
dep = dia['Diagnosis']                                    # dependent variable
```

```python
LR = LogisticRegression(max_iter=5)
LR.fit(ind, dep)
```

```python
age = int(input("Enter the Age: "))
Gender = input("Enter the Gender (Male/Female): ")
BMI = float(input("Enter the BMI: "))
Smoking = input("Enter the Smoking (Yes/No): ")
GeneticRisk = input("Enter the Genetic Risk (Yes/No): ")
PhysicalActivity = input("Enter the Physical Activity (Yes/No): ")
AlcoholIntake = input("Enter the Alcohol Intake (Yes/No): ")
CancerHistory = input("Enter the Cancer History (Yes/No): ")
```

```python
gender_num = 1 if Gender.lower() == 'male' else 0
smoking_num = 1 if Smoking.lower() == 'yes' else 0
genetic_risk_num = 1 if GeneticRisk.lower() == 'yes' else 0 # Assuming 'Yes' maps to 1, 'No' maps to 0 for simplicity given the input prompt.
physical_activity_num = 1 if PhysicalActivity.lower() == 'yes' else 0
alcohol_intake_num = 1 if AlcoholIntake.lower() == 'yes' else 0
cancer_history_num = 1 if CancerHistory.lower() == 'yes' else 0

pred = LR.predict([[age, gender_num, BMI, smoking_num, genetic_risk_num, physical_activity_num, alcohol_intake_num, cancer_history_num]])
pred1=le.fit_transform(pred)
print("Prediction:", pred1)
```

```python
print("Model accuracy on full data:", LR.score(ind, dep))
```
